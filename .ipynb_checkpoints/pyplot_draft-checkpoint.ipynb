```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y = np.random.randn(5)
x = np.arange(5)
fig, ax = plt.subplots(figsize=(4, 3))
ax.plot(x, y)
plt.show()
```

```python
data1, data2 = np.random.randn(2, 100)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color='b', linewidth=2, linestyle='--')
l2, = ax.plot(x, np.cumsum(data2), color='orange', linewidth=2)
l2.set_linestyle(':')
```

```python
type(l2), type(l2[0]), len(l2)
```

```python
def f(x, y):
    return x + y
x = np.linspace(0, 5, 10, dtype=np.float16)
y = np.linspace(5, 10, 10, dtype=np.float16)
x, x.shape, y, y.shape
```

```python
X, Y = np.meshgrid(x, y)
X[0], X.shape, Y[0], Y.shape
```

```python
Z = f(X, Y)
Z
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(Z, extent=[0, 5, 0, 5])
```
